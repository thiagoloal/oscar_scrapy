```python
import pandas as pd

movies = pd.read_csv('./database/movies_TMDB_PI2.csv')
movies = movies.set_index('movie')
movies = movies.sort_values('number_indications', ascending=False)
movies.head()
```

```python
movies.plot('indication_year', 'victories', s=movies['number_indications'].values*20,kind='scatter',figsize=(18, 16))
```

```python
import matplotlib.pyplot as plt
plt.show()
```
