```python
import pandas as pd
import numpy as np

movies = pd.read_csv('./database/movies_TMDB_PI2.csv')
movies = movies.set_index('movie')

del movies['Unnamed: 0']
```

```python
movies = movies.sort_values('number_indications', ascending=False)
movies = movies.loc[movies['tmdb_id'] <= 1]
```

```python
import requests
import json
api_key = 'aa043c2ae33192f63fcc150bb97816b7'


def find_movie(mv, yr):

    url = "https://api.themoviedb.org/3/search/movie?api_key="+api_key+"&language=en-US&query="+mv
    url= url+"&primary_release_year="+str(yr)
    payload = "{}"

    response = requests.request("GET", url, data=payload)
    data = json.loads(response.text)
    
    if len(data['results']) == 1:
        print(mv)
        return  data['results'][0]['id']
    
    elif data['results'] and data['total_results'] > 1:
        lmmv = []
        print(mv+' | '+str(len(data['results'])))
        for mvs in data['results']:
            if mvs['title'] == mv:
                lmmv.append(mvs)
                
        if len(lmmv) == 1:
            return  lmmv[0]['id']
        elif len(lmmv) > 1:
            return  False
        
        return  False
   
    return  False
```

```python
moviesdv=[]
print(len(movies.index.values))
for x in range(1,42):
    moviesdv.append(movies.iloc[int(len(movies.index.values)/41)*(x-1):int(len(movies.index.values)/41)*x,:].copy())
```

```python
for idx, df in enumerate(moviesdv):
    print(str(idx)+' | ===================================================================================')
    df['tmdb_id'] =  np.vectorize(find_movie)(df.index.values, df['indication_year'])
```
